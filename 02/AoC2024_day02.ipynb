```python
FILE_NAME = "day_02_input11.txt"

def load_data(file_name):
        file= open(file_name, "r")
        lines = file.readlines()
        return lines

def process(lines):
    safe = 0

    for line in lines:
        line = [int(x) for x in line.split()]

        distances = []
        for i in range(1, len(line)):
            distances.append(line[i-1] - line[i])
    
        if (all(val < 0 for val in distances) or all(val > 0 for val in distances)) and all(abs(val) < 4 for val in distances):
            safe += 1

    return safe

lines = load_data(FILE_NAME)
print(process(lines))
```
