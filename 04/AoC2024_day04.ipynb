```python
FILE_NAME = "day_04_input11.txt"

def load_data(file_name):
        file= open(file_name, "r")
        lines = file.read().splitlines()
        return lines

def process_line(lines):
    data = []
    for l in lines:
        line =[]
        for c in l:
            line.append(c)
        data.append(line)
    return data

def count(x, y, direction, word, data):
    ret = 0
    if y not in range(0, len(data[1])) or x not in range(0, len(data[1])):
        return ret
    
    if len(word) == 1 and data[y][x] == word[0]:
        return 1
    
    if data[y][x] == word[0]:
        if direction in ['B', '0'] :
            ret += count(x+1, y, "B", word[1:], data)
        if direction in ['BL', '0'] :
            ret += count(x+1, y-1, "BL", word[1:], data)
        if direction in ['L', '0'] :
            ret += count(x, y-1, "L", word[1:], data)
        if direction in ['LJ', '0'] :
            ret += count(x-1, y-1, "LJ", word[1:], data)
        if direction in ['J', '0'] :
            ret += count(x-1, y, "J", word[1:], data)
        if direction in ['JF', '0'] :
            ret += count(x-1, y+1, "JF", word[1:], data)
        if direction in ['F', '0'] :
            ret += count(x, y+1, "F", word[1:], data)
        if direction in ['FB', '0'] :
            ret += count(x+1, y+1, "FB", word[1:], data)
        return ret
    return ret

def process(lines):
    answer = 0
    word="XMAS"
    data =process_line(lines)
    for y in range(len(data)):
        for x in range(len(data)):
            answer += count(x, y, '0', word, data)
    return answer

lines = load_data(FILE_NAME)
print(process(lines))
```

```python
import re
import numpy as np

FILE_NAME = "day_04_input10.txt"

def load_data(file_name):
        file= open(file_name, "r")
        #text = file.read()

        lines = file.read().splitlines()
        print(lines)
        return lines

def process_line(lines):
    data = []
    
    for l in lines:
        line =[]
        for c in l:
            line.append(c)
        data.append(line)
    return data

def count(data):
    summa = 0
    lines = []
    # Reassemble each row and append to the list
    for row in data:
        lines.append(''.join(row))

    reg = r"XMAS"
    for i in lines:
        ttt = re.findall(reg, i)
        summa += len(ttt)
    print(summa)
    return summa


def process(lines):
    answer = 0
    data = np.array(process_line(lines))
    
    # simpla 
    answer += count(data)

    data_o = np.fliplr(data)

    #90 
    data_90 = np.rot90(data)
    answer += count(data_90)

    data_90_o = np.fliplr(data_90)

    # diag1
    data_diag1 = []
    for i in range(-int(data.shape[0])+1, int(data.shape[0])):
        data_diag1.append(data.diagonal(i))
    answer += count(data_diag1)

    #diag2
    data_diag2 = []
    for i in range(-int(data_90.shape[0])+1, int(data_90.shape[0])):
        data_diag2.append(data_90.diagonal(i))
    answer += count(data_diag2)

    #diag3
    data_diag3 = []
    for i in range(-int(data_o.shape[0])+1, int(data_o.shape[0])):
        data_diag3.append(data_o.diagonal(i))

    #diag3
    data_diag4 = []
    for i in range(-int(data_90_o.shape[0])+1, int(data_90_o.shape[0])):
        data_diag4.append(data_90_o.diagonal(i))

    answer += count(data_o)
    answer += count(data_diag3)
    answer += count(data_90_o)
    answer += count(data_diag4)




    return answer


lines = load_data(FILE_NAME)
print(process(lines))
```
