```python
!pip install parse
from parse import parse

FILENAME = "day_14_input11.txt"
TREE = "#"
EMPTY = "."
TREE_SEGMENT = TREE * 10
TIME_PART_ONE = 100
TIME_PART_TWO = 10_000

def parse_input(filename):
    with open(filename) as input_file:
        pattern = "p={p_x:d},{p_y:d} v={v_x:d},{v_y:d}"
        return [parse(pattern, row).named for row in input_file.read().split("\n")]


def create_image(bathroom, width, height):
    image = [[EMPTY for _ in range(width)] for _ in range(height)]
    for robot in bathroom:
        image[robot["p_y"]][robot["p_x"]] = TREE
    return "\n".join("".join(row) for row in image)


def simulate(bathroom, width, height, time):
    for robot in bathroom:
        robot["p_x"] = (robot["p_x"] + time * robot["v_x"]) % width
        robot["p_y"] = (robot["p_y"] + time * robot["v_y"]) % height
    return bathroom


def part_one(bathroom, width, height, time):
    bathroom = simulate(bathroom, width, height, time)
    mid_width, mid_height = width // 2, height //2
    upper_left, upper_right, lower_left, lower_right = 0, 0, 0, 0
    for robot in bathroom:
        p_x, p_y = robot["p_x"], robot["p_y"]
        if p_x < mid_width and p_y <mid_height:
            upper_left += 1
        if p_x < mid_width and p_y > mid_height:
            lower_left += 1
        if p_x > mid_width and p_y < mid_height:
            upper_right += 1
        if p_x > mid_width and p_y > mid_height:
            lower_right += 1
    return upper_left * upper_right * lower_left * lower_right


def part_two(bathroom, width, height, time):
    for t in range(time):
        new_bathroom = [{key: value for key, value in robot.items()} for robot in bathroom]
        new_bathroom = simulate(new_bathroom, width, height, t)
        image = create_image(new_bathroom, width, height)
        if TREE_SEGMENT in image:
            print(t, image, sep="\n")
            print()


def main():
    bathroom = parse_input(FILENAME)
    print(part_one(bathroom, 101, 103, TIME_PART_ONE))
    bathroom = parse_input(FILENAME)
    part_two(bathroom, 101, 103, TIME_PART_TWO)


if __name__ == "__main__":
    main()
```

```python
from re import findall
data = open("day_14_input11.txt").read()
W, H = 101, 103

robots = [[int(n) for n in findall(r"(-?\d+)", item)] for item in data.split("\n")]

def simulate(t):
    return [((sx + t*vx) % W, (sy + t*vy) % H) for (sx, sy, vx, vy) in robots]

from statistics import variance
bx, bxvar, by, byvar = 0, 10*100, 0, 10*1000
for t in range(max(W,H)):
    xs, ys = zip(*simulate(t))
    if (xvar := variance(xs)) < bxvar: bx, bxvar = t, xvar
    if (yvar := variance(ys)) < byvar: by, byvar = t, yvar
print("Part 2:", bx+((pow(W, -1, H)*(by-bx)) % H)*W)
```
